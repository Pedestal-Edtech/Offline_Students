```python
#  minion game
def minion_game(string):
    stuart = 0
    kevin = 0
    vowel = {'A','E','I','O','U'}
    s=input()

    for i in range(len(string)):
        if string[i] in vowel:
            kevin += len(string) - i
        else:
            stuart += len(string) - i
    if stuart > kevin:
        print(f"Stuart {stuart}")
    elif stuart == kevin:
        print("Draw")
    else:
        print(f"Kevin {kevin}")
```
