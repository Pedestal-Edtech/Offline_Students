```python
# inheritance
class car:
    def BMW(self):
        return "Running...."
class petrol(car):
    def engine(self):
        return "petrol car."
my_car=car()
print(my_car.BMW())
```

```python
mycar_ty=petrol()
print(mycar_ty.engine())
```

```python
class marks:
    def __init__(self):
        print( "failed...")
class subject(marks):
    def harshit1(self):
        print( "In maths")
my_marks=subject()
print(my_marks.harshit1())
```
