## nested dictionary

```python
car={'car1':{'brand':"toyota",'model':'innova','mf.year':2020},
     'car2':{"brand":'mahendra','model':'thar','year':2021},
    'car3':{'brand':'suzuki','model':'swift','year':2025}}
print(car)
print(car['car1'])

    
```

```python
print(car["car1"]['model'])
```

```python
car={'car1':{'brand':"toyota",'model':'innova','mf.year':2020},
     'car2':{"brand":'mahindra','model':'thar','year':2021},
    'car3':{'brand':'suzuki','model':'swift','year':2025}}
for i, obj in car.items():
    print(i)
    for j in obj:
        print(j + ":" , obj[j])
```
