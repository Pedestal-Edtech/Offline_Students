```python
def sum(a,b):
    return a+b

def subtraction(a,b):
    return a-b

def multiply(a,b):
    return a*b

def divide(a,b):
    return a/b

c=int(input("enter first number"))
d=int(input("enter second number"))

print("1. sum:")
print("2. subtraction:")
print("3. multiply:")
print("4. divide:")

choice=int(input("enter self choice: (1-4)"))

if choice==1:
    print("result=",sum(c,d))
elif choice==2:
    print("result=",subtraction(c,d))
elif choice==3:
    print("result=",multiply(c,d))
elif choice==4:
    print("result=",divide(c,d))
else:
    print("invalid input")
    
```

```python
# types of function
# 1. does'nt return value and no takes parameter
def sum():
    n1=int(input("enter value1"))
    n2=int(input("enter value2"))
    add=(n1+n2)
    print("sum=",add)
sum()
```

```python
# 2. not return a value but pass the parameter 
def add(n3,n4):
    sum=n3+n4
    print("add=",sum)
n3=int(input("enter the val1"))
n4=int(input("enter the val2"))
add(n3,n4)
```
