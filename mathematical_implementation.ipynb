```python
# program to add the elements of the list
ls1=[1,2,3,4]
sum=0
for i in range(len(ls1)):
    sum=sum+ls1[i]
print(sum)
    
```

```python
# program to print the elements of the list one by oneand print them column wise
ls2=[10,20,30,40]
sum=0
for i in range(len(ls2)):
    sum=sum+ls2[i]
    print(sum)
```

```python
ls3=[1,2,3,4]
ls3.reverse()
sub=ls3[0]
print(ls3)
for i in range(1,len(ls3)):
    sub=sub-ls3[i]
print(sub)
```
