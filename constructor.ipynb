```python
def index():
    ls2 = [3, 4]
    ls3 = [6]
    ls1 = [1, 2, ls2, 5, ls3]

    a = int(input("Enter any value: "))

    for i in range(len(ls1)):
        if isinstance(ls1[i], list:     
            for j in range(len(ls1[i])):
                if ls1[i][j] == a:
                    return (i, j)
        else:
            if ls1[i] == a:
                return (i,)

    return "Value not found"

print(index())
```
