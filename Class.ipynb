```python
class worker:
    def __init__(self,name,id,age):
        # print("rajesh")
        self.name=name
        self.id=id
        self.age=age
    def worker_details(self):
        print("worker details:")
        print("worker name:",self.name)
        print("worker id:",self.id)
        print("worker age:",self.age)
worker=worker("Harshit",100,21)
worker.worker_details()
```
