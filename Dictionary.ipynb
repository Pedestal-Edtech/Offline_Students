```python
dic2={'name':'Rajesh','age':'21','address':'Behror','degree':'B.tech'}
print(dic2)
print(type(dic2))
```

```python
# to pop the item from last into the dict
dic2.popitem()
print(dic2)
```

```python
# to pop any sepcific value form the dict
dic2.pop('age')
print(dic2)
```

```python
# to update any value in dict.
dic2.update({'class':'4th'})
print(dic2)
```

```python
# to get the value by key name from the dict
dic2.get('name')
```

```python
# to access all the values from the dict
dic2.values()
```

```python
# to get items with their key pair values and store them into list
dic2.items()
```

```python
# to get keys only from the dict
dic2.keys()
```

```python
# to get a specific value by key name 
dic2.setdefault('address')
```
