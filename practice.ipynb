```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y


def multiply(x, y):
    return x * y


print("Calculator")
print("1. Add")
print("2. Subtract")
print("3. Multiply")

method = int(input("Enter method (1-3): "))
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if method == 1:
    print("Result:", add(num1, num2))
elif method == 2:
    print("Result:", subtract(num1, num2))
elif method == 3:
    print("Result:", multiply(num1, num2))
else:
    print("Invalid function")
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y


def multiply(x, y):
    return x * y

def divide(x,y):
    return x / y 

def numinator(x,y):
    return x // y

def reminder(x, y):
    return x % y


print("Calculator")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. divide")
print("5. numinator")
print("6. reminder")

method = int(input("Enter method (1-6): "))
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print(num1)
print(num2)

if method == 1:
    print("Result in addition :", add(num1, num2))
elif method == 2:
    print("Result in substraction:", subtract(num1, num2))
elif method == 3:
    print("Result in multiplication:", multiply(num1, num2))
elif method == 4:
    print("result in divide:", divide(num1, num2))
elif method == 5:
    print("result is numinator:", numinator(num1 , num2))
elif method == 6:
    print("result is reminder:", reminder(num1,num2))
else:
    print("Invalid function")
```

```python
class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def show(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Marks: {self.marks}")

s1 = Student("chitra", 21, 87 )
s2 = Student("suhana", 21, 88)

s1.show()
s2.show()
```
