```python
## example for class, constructor, self, method

# creating class
class Student:
    def __init__(self, name, marks):       # constructor
        self.name  = name                  # save name into this object
        self.marks = marks                 # save marks into this object

    def topper(self):                                      # method
        print(self.name, "scored", self.marks)

    def grade(self):                                        # another method
        if self.marks >= 75:
            print(self.name, "- A")
        else:
            print(self.name, "- Fail")

# creating objects
s1 = Student("Suhana", 45)
s2 = Student("Chitra",   85)

# calling methods
s2.topper()    # Chitra scored 85
s2.grade()      # Chitra - A
s1.grade()      # Suhana - Fail
```

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def show_info(self):
        print("Brand :", self.brand)
        print("Speed :", self.speed)
    
    def is_fast(self):
        if self.speed >=100:
            print("fast car!")
        else:
            print("slow car")

    def describe(self):
         print(self.brand, "runs at", self.speed, "kmph")

c2 = Car("Toyota", 120)


c2.show_info()
c2.is_fast()
c2.describe()
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2*(self.length+self.width)
    
    def display(self):
        print("Area :", self.area())           
        print("Perimeter :", self.perimeter())

R1 = Rectangle(4,6)
R1.display()
    
```

```python
#question to swap string cases:
def swap_case(s):
    s=input()
    swaped_string=""
    for i in s:
        if i.isupper():
            swaped_string+=i.lower()
        elif i.islower():
            swaped_string+=i.upper()
        else:
            swaped_string += i
    return swaped_string
```
