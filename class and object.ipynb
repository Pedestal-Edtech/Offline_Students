## abstraction

```python
class car:
    def __init__(self):
        self.clutch=False
        self.acc=False
        self.brk=False
        print("car is stoped")
    def start(self):
        self.clutch=True
        self.acc=True
        print("car is start")
    def run(self):
        self.clutch=False
        self.brk=False
        self.acc=True
        print("car is running")
        
c=car()
c.start()
c.run()
        
    
```

## bank credit and debit program

```python
class account:
    def __init__(self, balance, account_no):
        self.balance=balance
        self.account_no=account_no
    def credit(self,amount):
        self.balance += amount
        print("Rs.",amount , "is credited in your account")
        print("total balance is ", self.balance)
    def debit(self,amount):
        self.balance -= amount
        print("Rs.", amount, "is debited from your account")
        print("total balance in your in account",self.balance)
a1=account(1000, 1234)
a1.credit(500)
a1.debit(200)
# print(a1.credit)
```
