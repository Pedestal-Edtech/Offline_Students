```python
# we create a list.
a=[1,2,3,4,8,6,5]
# we perform sort opertion to create values in seq.
a.sort()
# we perform reverse opertion now.
a.reverse()
# we use typecasting (set) to remove duplicates.
print(set(a))
print(a[1])
```

```python
# we taking a s variable to get values.
s=1,2,3,4
str(s)
# we convert string into list to use reverse opertion.
b=list(s)
b.reverse()
print(b)
```

```python
# remove duplicates using set.
u=[1,2,3,4,5,6]
# we convert list into set to remove duplicates.
u=set(u)
# again convert into list to use reverse opertion.
u=list(u)
u.reverse()
print(u)
```

```python
from collections import namedtuple
n,S=int(input()),namedtuple('S',input().split())
avg=sum((int(S._make(input().split()).MARKS) for _ in range(n))) / n
print("%.2f"%avg)
```
